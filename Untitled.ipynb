```python
import geopandas as gpd
import pandas as pd
```

```python
location = "https://raw.githubusercontent.com/nikkopante/data-camp/main/provinces/provinces.geojson"
gdf = gpd.read_file(location)
gdf.head()
```

```python
location = "./data_camp_registrants.csv"
registrants_df = pd.read_csv(location)
registrants_df.head(10)
```

```python
gdf = gdf.merge(registrants_df, left_on='ADM2_EN', right_on='Province')
gdf = gdf.drop(['ADM2_EN'], axis=1)

gdf.head()
```

```python
gdf.to_file("example.geojson", driver='GeoJSON')
```
